```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('train.csv')
```

```python
train_data.head()
```

```python
test_data = pd.read_csv('test.csv')
```

```python
test_data.head()
```

```python
train_ans = train_data['Survived']
train_data.drop('Survived',axis=1)
```

```python
train_data.drop(['Name','SibSp','Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'],axis=1,inplace=True)
#train_data = pd.get_dummies(train_data)
train_data.fillna(0.0)
```

```python
test_data.drop(['Name','SibSp','Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked'],axis=1,inplace=True)
#test_data = pd.get_dummies(test_data)
test_data.fillna(0.0)
```

```python
for df in [train_data,test_data]:
    df['Sex_binary']=df['Sex'].map({'male':1,'female':0})
train_data['Age'] = train_data['Age'].fillna(0)
test_data['Age'] = test_data['Age'].fillna(0)
features = ['Pclass','Age','Sex_binary']

```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=6, min_samples_leaf=6, min_samples_split=10)
```

```python
model.fit(train_data[features],train_ans)
```

```python
predictions = model.predict(test_data[features])
```

```python
sub = pd.DataFrame({'PassengerId':test_data['PassengerId'],'Survived':predictions})
```

```python
filename = 'Titanic.csv'
sub.to_csv(filename,index=False)
```

```python
from sklearn.metrics import accuracy_score
model_pred = model.predict(train_data[features])
print(accuracy_score(model_pred,train_ans))
```
