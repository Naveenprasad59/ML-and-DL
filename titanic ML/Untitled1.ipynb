```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('train.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.fillna(0.0)
```

```python
data.head()
```

```python
data = data.drop(['Name','PassengerId','Embarked','Cabin'],axis=1)
```

```python
data.head()
```

```python
y_train = data['Survived']
```

```python
data = data.drop(['Survived'],axis=1)
```

```python
data
```

```python
data['Sex'].replace('female',0,inplace=True)
data['Sex'].replace('male',1,inplace=True)
```

```python
data
```

```python
data = data.drop(['Ticket','SibSp','Parch'],axis=1)
```

```python
data
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(data,y_train)
```

```python
test = pd.read_csv('test.csv')
```

```python
test = test.drop(['Name','PassengerId','Embarked','Cabin','Ticket','SibSp','Parch'],axis=1)
```

```python
test
```

```python
test['Sex'].replace('female',0,inplace=True)
test['Sex'].replace('male',1,inplace=True)
```

```python
test = test.fillna(0.0)
```

```python
predict = clf.predict(test)
pas = pd.read_csv('test.csv')
```

```python
sub = pd.DataFrame({'PassengerId':pas['PassengerId'],'Survived':predict})
sub.to_csv('dec.csv',index=False)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters ={'criterion':['gini','entropy'],
              'splitter':['best','random'],
                  }
```

```python
from sklearn.metrics import f1_score,make_scorer
```

```python
scorer = make_scorer(f1_score)
```

```python
grid_obj = GridSearchCV(clf,parameters,scoring=scorer)
```

```python
grid_fit = grid_obj.fit(data,y_train)
best_clf = grid_fit.best_estimator_
```

```python
best_pred = best_clf.predict(test)
```

```python
sub = pd.DataFrame({'PassengerId':pas['PassengerId'],'Survived':best_pred})
sub.to_csv('decoptim.csv',index=False)
```
